## 2.0 Counting Values

```python
# Calcular a contagem de avaliações por 'user_id'
df_counts = df1['user_id'].value_counts().reset_index()

# Renomear as colunas
df_counts.columns = ['user_id', 'counts']

# Exibir o DataFrame
df_counts
```

```python
df_test = df_counts.loc[df_counts['counts'] != 1, :]
#df_test
df_test.loc[df_test['counts'] > 100, :].head()
```

## 2.1 Merging and Filtering 

```python
# Merging - Inner Join
df2 = pd.merge(df1, df_counts, on='user_id')

# Filtering 
df2_filtered = df2[df2['counts'] > 100]

# Visualization
df2_filtered.head()
```

```python
# Rows and Columns
df2_filtered.shape
```

## 2.2 Checking NA

```python
df2_filtered.isna().sum()
```

## 2.3 Drop Duplicates

```python
df2_filtered = df2_filtered.copy()

df2_filtered.drop_duplicates(['user_id','product_id'], inplace = True)
```

## 2.4 Removing Columns

```python
# Vamos excluir a variavel time_stamp porque não iremos utiliza-la
del df2_filtered['time_stamp']
```

## 2.5 Pivot

```python
products_pivot = df2_filtered.pivot_table(columns = 'user_id', index = 'product_id', values = 'ratings')

# Avaliar o arquivo transformado para PIVOT 
products_pivot.head(10)
```

```python
# Os valores que são nulos iremos preencher com ZERO
products_pivot.fillna(0, inplace = True)
products_pivot.head()
```

## 2.6 Sparse Matrix

```python
products_sparse = csr_matrix(products_pivot)
```

## 2.7 Training model

```python
# Criando e treinando o modelo preditivo para encontrar os vizinhos mais próximos
modelo = NearestNeighbors(algorithm = 'brute')
modelo.fit(products_sparse)
```

## 2.8 Testing

### 2.8.1 For product 0594481813

```python
distances, suggestions = modelo.kneighbors(products_pivot.filter(items=['0594481813'], axis=0).values.reshape(1, -1))

for i in range(len(suggestions)):
    for neighbor_index in suggestions[i]:
        neighbor_id = products_pivot.index[neighbor_index]
        if neighbor_id != '0594481813':
            print(neighbor_id)
```
